#### 6hrs/views.py

```python
from io import StringIO

from django.http import HttpResponse

depts_list = [
    {'no': 10, 'name': '', 'location': ''},
    {'no': 20, 'name': '', 'location': ''},
    {'no': 30, 'name': '', 'location': ''},
]


def index(request):
    output = StringIO()
    output.write('<html>\n')
    output.write('<head>\n')
    output.write('\t<meta charset="utf-8">\n')
    output.write('\t<title></title>')
    output.write('</head>\n')
    output.write('<body>\n')
    output.write('\t<h1></h1>\n')
    output.write('\t<hr>\n')
    output.write('\t<table>\n')
    output.write('\t\t<tr>\n')
    output.write('\t\t\t<th width=120></th>\n')
    output.write('\t\t\t<th width=180></th>\n')
    output.write('\t\t\t<th width=180></th>\n')
    output.write('\t\t</tr>\n')
    for dept in depts_list:
        output.write('\t\t<tr>\n')
        output.write(f'\t\t\t<td align=center>{dept["no"]}</td>\n')
        output.write(f'\t\t\t<td align=center>{dept["name"]}</td>\n')
        output.write(f'\t\t\t<td align=center>{dept["location"]}</td>\n')
        output.write('\t\t</tr>\n')
    output.write('\t</table>\n')
    output.write('</body>\n')
    output.write('</html>\n')
    return HttpResponse(output.getvalue())
```

```python
from django.shortcuts import render

depts_list = [
    {'no': 10, 'name': '', 'location': ''},
    {'no': 20, 'name': '', 'location': ''},
    {'no': 30, 'name': '', 'location': ''},
]


def index(request):
    return render(request, 'index.html', {'depts_list': depts_list})
```
